```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# טעינת הנתונים - זה השלב שמגדיר את המשתנה df
df = pd.read_csv('laptop_price - dataset.csv')

print("הנתונים נטענו והספריות מוכנות!")
```

```python
# יצירת שטח לגרף בגודל נוח
plt.figure(figsize=(12, 6))

# ציור הגרף - ציר ה-X הוא פשוט המספר הסידורי של הלפטופ, וציר ה-Y הוא המחיר
plt.plot(df['Price (Euro)'], color='blue', linewidth=1)

# הוספת כותרות (חשוב מאוד לציון!)
plt.title('Prices of All Laptops in the Dataset')
plt.xlabel('Laptop Index') # מספר הלפטופ בטבלה
plt.ylabel('Price (Euro)')

# הצגת הגרף
plt.show()
```

```python
# 1. חישוב ממוצע לכל חברה
# אנחנו אומרים: "קבץ לפי חברה, קח את עמודת המחיר, ותעשה ממוצע"
avg_prices = df.groupby('Company')['Price (Euro)'].mean()

# 2. סידור התוצאות מהיקר לזול כדי שיהיה לנו נוח לקרוא
avg_prices = avg_prices.sort_values(ascending=False)

# 3. הדפסת התוצאות
print("Average laptop price per company:")
print(avg_prices)

# 4. מציאת החברה הכי יקרה (הראשונה ברשימה אחרי המיון)
most_expensive = avg_prices.idxmax()
print(f"\nThe company with the most expensive laptops on average is: {most_expensive}")
```

```python
# הצגת כל הערכים הייחודיים בעמודה
print("Original Operating Systems in the data:")
print(df['OpSys'].unique())
```

```python
# יצירת מפה של "מה להחליף במה"
# צד שמאל: השם הישן, צד ימין: השם החדש והמאוחד
os_mapping = {
    'Windows 10': 'Windows',
    'Windows 7': 'Windows',
    'Windows 10 S': 'Windows',
    'macOS': 'Mac',
    'Mac OS X': 'Mac'
}

# ביצוע ההחלפה בתוך העמודה הקיימת
df['OpSys'] = df['OpSys'].replace(os_mapping)

# בדיקה שזה עבד - נדפיס שוב את הערכים הייחודיים
print("\nOperating Systems after cleaning:")
print(df['OpSys'].unique())
```

```python
# מקבלים רשימה של כל מערכות ההפעלה שיש לנו עכשיו (אחרי הניקוי)
unique_systems = df['OpSys'].unique()

# לולאה שעוברת מערכת-מערכת ומציירת גרף
for os_name in unique_systems:
    plt.figure(figsize=(8, 4)) # פתיחת דף חדש לכל גרף
    
    # סינון הנתונים: רק הלפטופים עם מערכת ההפעלה הנוכחית בלולאה
    subset = df[df['OpSys'] == os_name]
    
    # יצירת היסטוגרמה שמראה את התפלגות המחירים
    sns.histplot(subset['Price (Euro)'], kde=True, color='green')
    
    plt.title(f'Price Distribution for {os_name}')
    plt.xlabel('Price (Euro)')
    plt.ylabel('Count of Laptops')
    plt.show()
```

```python
# 1. גרף פיזור לבדיקת הקשר בין RAM למחיר
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['RAM (GB)'], y=df['Price (Euro)'])
plt.title('Relationship between RAM and Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (Euro)')
plt.show()

# 2. גרף קופסה לזיהוי חריגים במחיר
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Price (Euro)'])
plt.title('Detecting Price Outliers')
plt.show()
```

```python
# יצירת עמודה חדשה בשם 'Storage type'
# אנחנו לוקחים את עמודת Memory, ומפעילים עליה פונקציה (apply) 
# הפונקציה מחלקת את הטקסט למילים (split) ולוקחת את המילה האחרונה ([-1])
df['Storage type'] = df['Memory'].apply(lambda x: x.split()[-1])

# בואי נראה את התוצאה עבור 10 המחשבים הראשונים
print("New Storage type column:")
print(df[['Memory', 'Storage type']].head(10))
```
