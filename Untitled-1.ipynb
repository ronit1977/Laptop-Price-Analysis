## Step 1: Data Loading and Validation
We import the libraries and check if the dataset is loaded correctly with valid numeric prices.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load data and validate that the price column is numeric
df = pd.read_csv("laptop_price - dataset.csv")
price_col = "Price (Euro)"

# Convert column to numeric, invalid entries become 'NaN' (Not a Number)
df[price_col] = pd.to_numeric(df[price_col], errors="coerce")

# Check if there are any invalid or missing prices
missing_prices = df[price_col].isna().sum()

if missing_prices > 0:
    print(f"Warning: Found {missing_prices} invalid price entries.")
    # Option: remove rows with invalid prices
    df = df.dropna(subset=[price_col])
    print("Cleaned the dataset by removing invalid prices.")
else:
    print("Validation Success: All prices are numeric and valid!")
# Previewing the data to ensure it was loaded correctly
print("Data successfully loaded. Here are the first 5 rows:")
df.head()
```

## Step 2: Global Price Overview
Visualizing the distribution of all laptop prices in the dataset.

```python
# Task: Plot the price of all the laptops
plt.figure(figsize=(12, 6))
plt.plot(df['Price (Euro)'], color='royalblue', linewidth=1)

# Adhering to guidelines: Adding plot title, axes titles, and units
plt.title('Distribution of All Laptop Prices', fontsize=14)
plt.xlabel('Laptop Entry Index', fontsize=12)
plt.ylabel('Price (Euro)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Step 3: Company Analysis
Calculating the average price for each brand to identify the most expensive company.

```python
# Task: Find average laptop price for each company and the most expensive one
avg_price_per_company = df.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False)

print("Average Price per Company (in Euro):")
print(avg_price_per_company)

# Identifying the company with the highest average price
most_expensive = avg_price_per_company.idxmax()
print(f"\nResult: The company with the most expensive laptops on average is {most_expensive}.")#
```

## Step 4: Operating System Analysis & Cleaning
Standardizing OS names and plotting price distributions for each unique system.

```python
# Task: Find different types of Operating Systems and fix them to be uniform
print("Original Operating Systems in the data:")
print(df['OpSys'].unique())

# Mapping similar systems to uniform names
os_mapping = {
    'Windows 10': 'Windows',
    'Windows 7': 'Windows',
    'Windows 10 S': 'Windows',
    'macOS': 'Mac',
    'Mac OS X': 'Mac'
}
df['OpSys'] = df['OpSys'].replace(os_mapping)

print("\nOperating Systems after cleaning (Uniform names):")
print(df['OpSys'].unique())
```

## Step 5: RAM vs. Price & Outlier Detection
Examining how RAM affects the cost and identifying statistical outliers in the price data.

```python
# Task: Plot the price distribution for each unique operating system
unique_os_list = df['OpSys'].unique()

for os_name in unique_os_list:
    plt.figure(figsize=(8, 4))
    # Filtering data for the specific OS
    os_data = df[df['OpSys'] == os_name]
    
    # Plotting histogram with Kernel Density Estimate (KDE)
    sns.histplot(os_data['Price (Euro)'], kde=True, color='seagreen')
    
    # Adding titles and units
    plt.title(f'Price Distribution for {os_name} Operating System', fontsize=13)
    plt.xlabel('Price (Euro)', fontsize=11)
    plt.ylabel('Count of Laptops', fontsize=11)
    plt.show()
```

## Step 6: Feature Engineering & Bonus Analysis
Extracting storage types and analyzing their impact on laptop pricing.

```python
# Task: Create a new column "Storage type" extracting from "Memory"
# Logic: Split the string and take the last element (e.g., 'SSD' or 'HDD')
df['Storage type'] = df['Memory'].apply(lambda x: x.split()[-1])

# Showing the transformation
print("Updated Data (First 10 rows):")
print(df[['Memory', 'Storage type']].head(10))
```

```python
# Task 5: Relationship between RAM and Computer Price & Outlier Analysis


# 1. Visualizing the relationship using a Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['RAM (GB)'], y=df['Price (Euro)'], alpha=0.7, color='dodgerblue')

# Adding required titles, labels, and units
plt.title('Relationship between RAM Capacity and Laptop Price', fontsize=14)
plt.xlabel('RAM (GB)', fontsize=12)
plt.ylabel('Price (Euro)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Finding: There is a clear positive correlation; as RAM increases, the price generally rises.

# 2. Checking for outliers using a Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Price (Euro)'], color='orange')

# Adding required titles, labels, and units
plt.title('Detection of Price Outliers', fontsize=14)
plt.xlabel('Price (Euro)', fontsize=12)
plt.show()



```

Q: What would be considered as an outlier? 
A: An outlier is a data point that is significantly different from the 
   remaining data (e.g., an extremely expensive laptop that doesn't 
   follow the typical price-to-spec trend).

Q: How will you detect it?
A: 1. Visually: Using the Boxplot above. Points (dots) appearing outside 
      the 'whiskers' of the plot are statistically defined as outliers.
   2. Mathematically: By using the Interquartile Range (IQR) method. 
      Any value above [Q3 + 1.5 * IQR] or below [Q1 - 1.5 * IQR] is 
      flagged as an outlier.

```python
# Bonus: Average Price by Storage Type
# This analysis shows which storage technology is generally more expensive
plt.figure(figsize=(10, 5))
bonus_analysis = df.groupby('Storage type')['Price (Euro)'].mean().sort_values()
bonus_analysis.plot(kind='barh', color='purple')

plt.title('Bonus: Average Price (Euro) by Storage Technology', fontsize=13)
plt.xlabel('Average Price (Euro)', fontsize=11)
plt.ylabel('Storage Type', fontsize=11)
plt.show()
```
