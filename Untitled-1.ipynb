# MiniProject: Laptop Price Data Analysis



## Step 1: Data Loading and Validation
We import the libraries and check if the dataset is loaded correctly with valid numeric prices.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("laptop_price - dataset.csv")
price_col = "Price (Euro)"

# Data Validation: Convert price column to numeric, turning invalid entries into NaN
df[price_col] = pd.to_numeric(df[price_col], errors="coerce")

# Check and clean missing/invalid prices
missing_prices = df[price_col].isna().sum()

if missing_prices > 0:
    print(f"Warning: Found {missing_prices} invalid price entries. Removing them...")
    df = df.dropna(subset=[price_col])
    print("Cleanup complete.")
else:
    print("Validation Success: All prices are valid numeric values.")

# Display general info about the dataset (Columns, Data Types, Non-Null counts)
print("\nDataset Overview:")
df.info()

# Preview the first 5 rows
print("\nFirst 5 rows of the dataset:")
df.head()
```

## Step 2: Global Price Overview
Visualizing the distribution of all laptop prices in the dataset.

```python
# Task: Plot the price distribution of all laptops with a trend line (KDE)
plt.figure(figsize=(12, 6))

# Adding kde=True will draw the smooth "trend" line over the bars
sns.histplot(df['Price (Euro)'], bins=30, color='royalblue', edgecolor='white', kde=True)

# Adding plot title, axes titles, and units as required by the guidelines
plt.title('Distribution of All Laptop Prices with Trend Line', fontsize=14)
plt.xlabel('Price (Euro)', fontsize=12)
plt.ylabel('Number of Laptops', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

## Step 3: Company Analysis
Calculating the average price for each brand to identify the most expensive company.

```python
# Task: Find average laptop price for each company and the most expensive one
# Grouping by Company and calculating the mean price
avg_price_per_company = df.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False)

print("Average Price per Company (in Euro):")
print(avg_price_per_company)

# Identifying the company name and its average price
most_expensive_name = avg_price_per_company.idxmax()
most_expensive_value = avg_price_per_company.max()

# Final result including both the company name and the calculated average price
print(f"\nResult: The company with the most expensive laptops on average is {most_expensive_name} "
      f"with an average price of {most_expensive_value:.2f} Euro.")
# Visualizing the average price per company for better comparison
plt.figure(figsize=(12, 5))
avg_price_per_company.plot(kind='bar', color='green', edgecolor='black')
plt.title('Average Price per Company', fontsize=14)
plt.ylabel('Price (Euro)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## Step 4: Operating System Analysis & Cleaning
Standardizing OS names and plotting price distributions for each unique system.

```python
# Task: Find different types of Operating Systems and fix them to be uniform
# Showing before and after to verify the cleaning process

#  Create a simple mapping for names that need to be changed
# Names NOT in this list (like 'No OS' or 'Linux') will remain unchanged
os_fix = {
    'Windows 10': 'Windows',
    'Windows 7': 'Windows',
    'Windows 10 S': 'Windows',
    'Mac OS X': 'macOS'
}

# Perform the fix
df['OpSys'] = df['OpSys'].replace(os_fix)

# 3. Show the unique names AFTER the fix
print("Unique OS after fix:", df['OpSys'].unique())
```

```python
# Task: Plot the price distribution for each unique operating system
# Using sns.displot to create a grid of plots (Faceting) instead of a loop

# Creating the distribution plots for each Operating System
graph_of_op_price = sns.displot(
    data=df, 
    x="Price (Euro)", 
    col="OpSys", 
    hue="OpSys", 
    col_wrap=3, 
    kde=True, 
    height=4, 
    facet_kws={'sharex': False, 'sharey': False}
)

# Setting uniform axis labels
graph_of_op_price.set_axis_labels("Price (Euro)", "Number of Laptops") 

# Adding a main title for the entire figure
graph_of_op_price.fig.suptitle("Distribution of Laptop Prices (Euro) by Operating System", y=1.05, fontsize=16)

plt.show()
```

## Step 5: RAM vs. Price & Outlier Detection
Examining how RAM affects the cost and identifying statistical outliers in the price data.

```python
import numpy as np

# 1. Statistical Detection (The "How")
Q1 = df["Price (Euro)"].quantile(0.25)
Q3 = df["Price (Euro)"].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# 2. Filter the data
df_clean = df[df["Price (Euro)"] <= upper_bound]

# 3. Professional Visualization (Using your lmplot idea)
# This shows the average price for each RAM size clearly
plot = sns.lmplot(
    x="RAM (GB)", 
    y="Price (Euro)", 
    data=df_clean, 
    x_estimator=np.mean, 
    height=6, aspect=1.5,
    line_kws={'color': 'red'},
    scatter_kws={'color': 'blue'}
)

# 4. Mandatory Labels and Units (From PDF requirements)
plot.set_axis_labels("RAM (GB)", "Average Price (Euro)")
plt.title("Relationship between RAM and Average Price (Cleaned Data)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(f"Outliers detected above: {upper_bound:.2f} Euro")
print(f"Correlation (Clean Data): {df_clean['RAM (GB)'].corr(df_clean['Price (Euro)']):.3f}")
```

### Outlier Analysis and Methodology

**1. Visual Detection**
In the initial scatter plot, we can visually identify outliers as data points that deviate significantly from the general trend (the regression line). These represent laptops that are exceptionally expensive relative to their specifications, creating "noise" in the analysis.

**2. Why use the IQR (Interquartile Range) method?**
To address this, I chose the **IQR method** rather than manually removing specific groups (such as all 64GB RAM laptops).
* **Objectivity:** This is a statistical, data-driven approach. It defines outliers based on the mathematical distribution of the data ($Q3 + 1.5 \times IQR$) rather than arbitrary decisions.
* **Accuracy:** It ensures we only remove truly extreme values while keeping high-end laptops that fit the general price trend.

**Conclusion**
By filtering these statistical outliers, we reveal a clearer linear relationship between RAM and Price, resulting in a more accurate correlation coefficient.

## Step 6: Feature Engineering - Extracting Storage Type
Parsing the complex text in the 'Memory' column to create a clean, usable column for storage technologies.

```python
# Task: Create a new column "Storage type" by extracting it from "Memory"
# Logic: Split the string and keep only words that represent the storage technology (not numbers)

def memory_to_storage_type(memory):
    # Ensure the input is treated as a string
    memory = str(memory)
    storage_parts = []
    
    # Split the memory description into separate words
    words = memory.split()
    
    for word in words:
        # If the word does not start with a number (like 128, 512, 1.0), it's the storage type
        if word and not word[0].isnumeric():
            storage_parts.append(word)
            
    # Join the words back (handles cases like "SSD", "HDD", or "Flash Storage")
    return " ".join(storage_parts)

# Applying the function to create the new column
df["Storage type"] = df["Memory"].apply(memory_to_storage_type)

# Showing only the relevant columns to compare the original vs the extracted result
print("Storage Type Extraction Results (Original vs New):")
print(df[['Memory', 'Storage type']].head(15))

```
